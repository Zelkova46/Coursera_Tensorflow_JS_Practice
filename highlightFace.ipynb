```python
import skimage
import numpy as np
from numpy import asarray
from skimage import io
from matplotlib import pyplot as plt
from skimage import img_as_float
from skimage import img_as_ubyte
import glob
import cv2
```

```python
list_gambar = []
list_gambar_gray = []
path = "/content/datasets/*.*"
namaFile = []
for file in sorted(glob.glob(path)):
  print(file) #anggap aja file kaya path tempat si gambar 

  img = io.imread(file) 
  namaFile.append(file)
  img_gray = io.imread(file, as_gray = True)

  list_gambar.append(img)
  list_gambar_gray.append(img_gray)
```

```python
for x in range(len(list_gambar)):

  # io.imshow(list_gambar_gray[x])


  # plt.imshow(list_gambar_gray[x], cmap = "jet") 

  # plt.figure()
  plt.subplot(121)
  plt.imshow(list_gambar_gray[x], cmap="gray")
  plt.title(namaFile[x])
  plt.subplot(122)
  plt.imshow(list_gambar_gray[x], cmap="jet")
  plt.title('Processed Image')
  plt.show()
```
